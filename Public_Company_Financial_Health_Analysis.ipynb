```python
import yfinance as yf
import pandas as pd
```

```python
tickers = ["JPM", "BAC", "C", "WFC", "GS"]
```

```python
income_statements = {}

for ticker in tickers:
    stock = yf.Ticker(ticker)
    income_statements[ticker] = stock.financials
```

## Save Income Statements

```python
from pathlib import Path

raw_dir = Path("Raw_Data")
raw_dir.mkdir(exist_ok=True)

for ticker, df in income_statements.items():
    df.to_csv(raw_dir / f"{ticker}_income_statement.csv")
```

## Save Balance Sheets and Cashflow Statements

```python
balance_sheets = {}

for ticker in tickers:
    stock = yf.Ticker(ticker)
    balance_sheets[ticker] = stock.balance_sheet
raw_dir = Path("Raw_Data")
raw_dir.mkdir(exist_ok=True)
for ticker, df in balance_sheets.items():
    df.to_csv(raw_dir / f"{ticker}_balance_sheet.csv")
    
cash_flows = {}

for ticker in tickers:
    stock = yf.Ticker(ticker)
    cash_flows[ticker] = stock.cashflow

raw_dir = Path("Raw_Data")
raw_dir.mkdir(exist_ok=True)
for ticker, df in cash_flows.items():
    df.to_csv(raw_dir / f"{ticker}_cash_flows.csv")
```
